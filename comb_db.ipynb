```python
import pandas as pd
```

```python
email_df = pd.read_csv("enron_spam_cleaned.csv")
sms_df = pd.read_csv("sms_spam_cleaned.csv")
```

```python
print("Email shape:", email_df.shape)
print("SMS shape:", sms_df.shape)

email_df.head(), sms_df.head()
```

```python
combined_df = pd.concat([email_df, sms_df], axis=0)
```

```python
# Shuffle to mix email & SMS rows
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
print("Combined shape:", combined_df.shape)
# counts the total number of spam and ham
print(combined_df['label'].value_counts())
```

```python
combined_df.to_csv("spam_shield_combined.csv", index=False)

print("Combined dataset saved as spam_shield_combined.csv")
```
