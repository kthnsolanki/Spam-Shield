```python
import pandas as pd
import re
```

```python
# Load SMS dataset
sms = pd.read_csv("spam.csv", encoding="latin-1")
```

```python
# Keep only required columns
sms = sms[['v1', 'v2']]
```

```python
# Rename columns to match Enron format
sms = sms.rename(columns={'v1': 'label', 'v2': 'text'})
```

```python
# Drop missing values
sms = sms.dropna(subset=['text', 'label'])
```

```python
# Normalize labels
sms['label'] = sms['label'].str.lower().str.strip()
```

```python
# Basic text cleaning function
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\n|\r', ' ', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

sms['text_clean'] = sms['text'].apply(clean_text)
```
