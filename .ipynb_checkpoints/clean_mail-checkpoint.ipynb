```python
import pandas as pd
import re
```

```python
df = pd.read_csv("enron_spam_data.csv")
```

```python
df.columns = [c.lower().strip().replace(" ", "_") for c in df.columns]
```

```python
df = df.rename(columns={
    'message_id': 'id',
    'subject': 'subject',
    'message': 'text',
    'spam/ham': 'label',
    'date': 'date'
})
```

```python
# Drop missing text or label
df = df.dropna(subset=['text', 'label'])
```

```python
# Normalize labels
df['label'] = df['label'].str.lower().str.strip()
```

```python
df = df.drop_duplicates(subset=['text'])
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\n|\r', ' ', text)                 # remove line breaks
    text = re.sub(r'http\S+|www\S+', '', text)        # remove URLs
    text = re.sub(r'\bfrom:.*', '', text)             # remove headers
    text = re.sub(r'\bto:.*', '', text)
    text = re.sub(r'\bcc:.*', '', text)
    text = re.sub(r'-{2,}', ' ', text)                # remove ---- separators
    text = re.sub(r'\d{3}[-.\s]??\d{3}[-.\s]??\d{4}', '', text)  # phone numbers
    text = re.sub(r'[^a-z\s]', ' ', text)             # remove punctuation & numbers
    text = re.sub(r'\s+', ' ', text).strip()          # extra spaces
    return text
```

```python
df['text_clean'] = df['text'].apply(clean_text)
```

```python
# Keep only what ML needs
final_df = df[['text_clean', 'label']]
```

```python
final_df.to_csv("enron_spam_cleaned.csv", index=False)
print("Cleaned file saved as enron_spam_cleaned.csv")
print(final_df.head())
```
