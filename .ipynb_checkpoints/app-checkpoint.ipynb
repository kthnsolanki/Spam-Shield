```python
import streamlit as st
from model_util import predict_msg
```

```python
st.set_page_config(
    page_title="Spam Shield",
    layout="centered"
)
```

```python
st.title("Spam Shield")
st.write("Check whether a message is Spam or Safe")
```

```python
user_input = st.text_area(
    "Enter SMS or Email:",
    height=150
)
```

```python
if st.button("Check Message"):
    if user_input.strip() == "":
        st.warning("Please enter some text")
    else:
        result = predict_msg(user_input)

        if result == "spam":
            st.error("Alert SPAM message detected")
        else:
            st.success("This message is NOT spam")
```
